```python
LX=128

mode='fa'
value_name = mode.capitalize() if mode=='fa' else mode

base_path = f'z:/mips/{mode}_study/LX{LX}' #/output_Fa_{Fa} #/exp_{exp}'
outpath = os.path.join(base_path, 'positions')

folder_list_unsorted = glob.glob(os.path.join(base_path, f'output_{value_name}_*'))
act_list_unsorted = [float(x.split(f'{value_name}_')[-1]) for x in folder_list_unsorted]
folder_list, act_list = zip(*sorted(zip(folder_list_unsorted, act_list_unsorted), key=lambda x: x[1]))

#data=read_mips_files(filepath, max_timesteps=50, return_velocity=True, verbose=True)
folder_list
```

```python
# load defect_list from pkl
exp=0



# do scatter plot of defects for first 5 frames
act_to_plot = [0, 2, 4, 6, 20,22,28,30]
ffidx=100
fjump=10
nf=1
ncols = len(act_to_plot)
fig, ax = plt.subplots(ncols=ncols, nrows=1, figsize=(16,4))
fig.dpi=220
for i, act in enumerate(act_to_plot):
    exp=0
    axx = ax
    
    save_path = os.path.join(outpath, f'{value_name}_{act}',f'exp_{exp}')
    with open(os.path.join(save_path, f'defect_positions.pkl'), 'rb') as f  :
        top_defects = pickle.load(f)

  #  axx[0].set_ylabel(f'rho {act}', fontsize=16)
    for j in range(nf):
        j=i
        fidx = ffidx + j * fjump
        data = top_defects[fidx]

        axx[j].scatter(data[:,0], data[:,1], s=10, c='blue',alpha=.3)
        axx[j].set_xlim(0, LX)
        axx[j].set_ylim(0, LX)
        axx[j].set_title(f'rho {act}')
        # turn of ticks and ticklabels
        axx[j].set_xticks([])
        axx[j].set_yticks([])
        axx[j].set_aspect('equal', adjustable='box')
fig.tight_layout()
```
